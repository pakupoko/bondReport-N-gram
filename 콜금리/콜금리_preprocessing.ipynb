```python
import pandas as pd
```

```python
df=pd.read_excel('call.xlsx', encoding='utf-8')
```

```python
df=df.set_index(['date'])
```

```python
new_df=df.loc[:'2005-01-01']
```

```python
from datetime import timedelta
updown=[]
for idx in df.index:
        time = idx - pd.DateOffset(months=1)
        
        while True:
            if idx.year <= 2004 : break
            
            try:
                if df.loc[idx][0]>df.loc[time][0] : updown.append('상승')
                elif df.loc[idx][0]<df.loc[time][0]: updown.append('하락')
                else : updown.append('0')
                break
            except:
                time-=timedelta(1)
                continue
```

```python
new_df['UpDown']=updown
```
