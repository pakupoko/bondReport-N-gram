```python
import pandas as pd
```

```python
import sys
import csv
maxInt = sys.maxsize

while True:
    # decrease the maxInt value by factor 10 
    # as long as the OverflowError occurs.

    try:
        csv.field_size_limit(maxInt)
        break
    except OverflowError:
        maxInt = int(maxInt/10)
```

```python
df=pd.read_csv('bond_report2008_2017.csv', engine='python', encoding='utf-8')
```

```python
from ekonlpy.sentiment import MPCK

def text2ngram(text):
    try:
        mpck = MPCK()
        bef_tokens = mpck.tokenize(text)
        ngrams = mpck.ngramize(bef_tokens)

        return bef_tokens, ngrams
    except:
        return "NaN", 'NaN'
```

```python
token_list=[]
ngram_list=[]
for idx, text in enumerate(df['텍스트']):
    tokens, ngram = text2ngram(text)
    token_list.append(tokens)
    ngram_list.append(ngram)
    print(f'{df["날짜"][idx]} 완료')
```

```python
df['n-gram']=ngram_list
```

```python
df.to_csv('bonReport_ngram.csv',encoding='utf-8', index=False)
```
