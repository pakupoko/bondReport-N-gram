```python
import pandas as pd
```

```python
#csv 리딩 용량 확대
import sys
import csv
maxInt = sys.maxsize

while True:
    # decrease the maxInt value by factor 10 
    # as long as the OverflowError occurs.

    try:
        csv.field_size_limit(maxInt)
        break
    except OverflowError:
        maxInt = int(maxInt/10)
```

```python
df_bond = pd.read_csv('bondReport_ngram.csv', encoding='utf-8')
df_bond
```

```python
df_bond=df_bond.drop(['Unnamed: 0'], axis=1)
```

```python
df_bond
```

```python
df_call=pd.read_excel('콜금리_updown.xlsx', encoding='utf-8')
```

```python
df_call['date']=pd.to_datetime(df_call['date'])
df_call=df_call.set_index('date')
```

```python
df_bngram=df_bond.loc[:,['날짜','n-gram']]
```

```python
df_bngram['날짜']=pd.to_datetime(df_bngram['날짜'])
```

```python
updown=[]
for date in df_bngram['날짜']:
    try:
        updown.append(df_call.loc[date][1])
    except:
        updown.append('NaN')
```

```python
df_bngram['UpDown']=updown
```

```python
df_bngram
```

```python
df_bngram.to_csv('bond_Labeling.csv', index='False', encoding='utf-8')
```
